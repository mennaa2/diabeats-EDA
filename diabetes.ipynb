# Outliers Treatment

```python
Q1 = data['Glucose'].quantile(0.25)
Q3 = data['Glucose'].quantile(0.75)

IQR = Q3 - Q1

print("Q1 (25th percentile):", Q1)
print("Q3 (75th percentile):", Q3)
print("IQR:", IQR)

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print("Lower Limit:", lower_limit)
print("Upper Limit:", upper_limit)

data = data[(data['Glucose'] >= lower_limit) & (data['Glucose'] <= upper_limit)]

print("Rows remaining after removing outliers:", data.shape[0])

data.boxplot(column='Glucose', grid=False)
plt.title('Glucose (After IQR Outlier Removal)')
plt.ylabel('Glucose')
plt.show()
```
